```python
list_automobile = [['volvo',30000],['oppo',40000],['samsung',50000],['apple',60000]]
df_mob1=pd.DataFrame(list_automobile,columns=["vname","price"])
df_mob1
```

```python
list_automobile = [[1,'volvo',30],[2,'oppo',40],[3,'samsung',50],[4,'apple',60]]
df_mob3=pd.DataFrame(list_automobile,columns=["index","vname","size"])
df_mob3
```

```python
list_automobile = [[1,'vivo',35000],[2,'oneplus',45000],[3,'realme',57000],[4,'apple',60000]]
df_mob2=pd.DataFrame(list_automobile,columns=["index","vnames","price"])
df_mob2
```

# joins

```python
df_mob2.set_index("index").join(df_mob3.set_index("index"),on='index',how='inner')
```

# Merge

```python
pd.merge(df_mob2,df_mob3, on = 'index')
```

```python
pd.merge(df_mob2,df_mob3, on = 'index',how = 'outer')
```

```python
pd.merge(df_mob2,df_mob3, on = 'index',how = 'right')
```

```python
pd.merge(df_mob2,df_mob3, on = 'index',how = 'left')
```

```python
pd.merge(df_mob2,df_mob3,left_index = True, right_index = True)
```

```python
df_melt = df_mob2.melt(id_vars=['index'])
df_melt
```

```python
df_melt = df_mob2.melt(id_vars=['index'] , value_vars='vname')
df_melt
```

# pivot

```python
pd.pivot_table(df_mob2,index=['vnames'], values=['price'])
```

# cross tables

```python
pd.crosstab(df_mob2.vnames,df_mob3.size,rownames=['mobile'],colnames=['size'])
```
