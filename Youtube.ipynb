```python
mode1=df["category_id"].mode().values[0]
df["category_id"]=df["category_id"].replace(np.nan,mode1)
```

```python
df.isnull().sum()
```

```python
#droped these columns as all values are unique
df.drop("channel_title",axis=1,inplace=True)
df.drop("title",axis=1,inplace=True)
df.drop("tags",axis=1,inplace=True)
df.drop("description",axis=1,inplace=True)
```

```python
mean1=df["subscriber"].mean()
df["subscriber"]=df["subscriber"].replace(np.nan,mean1)
```

```python
df.isnull().sum()
```

```python
mean2=df["Trend_day_count"].mean()
df["Trend_day_count"]=df["Trend_day_count"].replace(np.nan,mean2)
```

```python
mode2=df["Tag_count"].mode().values[0]
df["Tag_count"]=df["Tag_count"].replace(np.nan,mode2)
```

```python
mode3=df["Trend_tag_count"].mode().values[0]
df["Trend_tag_count"]=df["Trend_tag_count"].replace(np.nan,mode3)
```

```python
mode4=df["tag appered in title"].mode().values[0]
df["tag appered in title"]=df["tag appered in title"].replace(np.nan,mode4)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False)
```

```python
# dropping unnamed :17 and unnamed :18 columns because it has 3197 null values
```

```python
df.drop("Unnamed: 17",axis=1,inplace=True)
df.drop("Unnamed: 18",axis=1,inplace=True)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False)
```

# Encoding the categorical columns

```python
Comment_disabled=pd.get_dummies(df["comment_disabled"])
Comment_disabled.head()
```

```python
Comment_disabled=pd.get_dummies(df["comment_disabled"],drop_first="True")
Comment_disabled.head()
```

```python
Comment_disabled=pd.get_dummies(df["comment_disabled"],drop_first="True")
Comment_disabled.head()
```

```python
likedislike=pd.get_dummies(df["like dislike disabled"])
likedislike.head()
```

```python
likedislike=pd.get_dummies(df["like dislike disabled"],drop_first="True")
likedislike.head()
```

```python
tagintitle=pd.get_dummies(df["tag appered in title"])
tagintitle.head()
```

# Concatenating Values

```python
df.head()
```

```python
df=pd.concat([df,Comment_disabled],axis=1)
df.drop(["FALSE"],axis=1,inplace=True)
df.drop(["TRUE"],axis=1,inplace=True)
df.drop(["comment_disabled"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df=pd.concat([df,Comment_disabled],axis=1)
```

```python
df.rename(columns = {'TRUE':'Comment_disabled_TRUE'}, inplace = True)
df.drop(["FALSE"],axis=1,inplace=True)
```

```python
df=pd.concat([df,likedislike],axis=1)
```

```python
df.drop(["FALSE"],axis=1,inplace=True)
df.rename(columns = {'TRUE':'likedislike_TRUE'}, inplace = True)
```

```python
df.drop(["like dislike disabled"],axis=1,inplace=True)
```

```python
df=pd.concat([df,tagintitle],axis=1)
df.rename(columns = {'True':'tagintitle_TRUE'}, inplace = True)
```
